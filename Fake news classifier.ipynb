```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
import pandas as pd
```

```python
fake=pd.read_csv("fake.csv")
```

```python
true=pd.read_csv("true.csv")
```

```python
fake["target"]=0
true["target"]=1
```

```python
fake
```

```python
true
```

```python
data=pd.concat([fake,true],axis=0)
```

```python
data=data.reset_index(drop=True)
```

```python
data
```

```python
data=data.drop(["subject","title","date"],axis=1)
```

```python
data
```

```python
from nltk.tokenize import word_tokenize
```

```python
data["text"]=data["text"].apply(word_tokenize)
```

```python
data
```

```python
from nltk.stem.snowball import SnowballStemmer
stem=SnowballStemmer("english")
```

```python
def stem_it(text):
    return(stem.stem(word) for word in text)
```

```python
data["text"]=data["text"].apply(stem_it)
```

```python
print(data.head(10))
```

```python
def stop_it(t):
    dt=[word for word in t if len(word)>2]
    return dt
```

```python
data["text"]=data["text"].apply(stop_it)
```

```python
data["text"]=data["text"].apply(" ".join)
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data["text"],data["target"],test_size=0.25)
```

```python
display(x_train.head())
print("\n")
display(y_train.head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
my_tfidf=TfidfVectorizer(max_df=0.7)

tfidf_train=my_tfidf.fit_transform(x_train)
tfidf_test=my_tfidf.transform(x_test)
```

```python
print(tfidf_train)
```

# Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
model_1=LogisticRegression(max_iter=900)
model_1.fit(tfidf_train,y_train)
prediction_1=model_1.predict(tfidf_test)
pred_score=accuracy_score(y_test,prediction_1)
print(pred_score*100)
```

# Passive Agressive Classifier Model


```python
from sklearn.linear_model import PassiveAggressiveClassifier
model_2=PassiveAggressiveClassifier(max_iter=60)
model_2.fit(tfidf_train,y_train)
prediction_2=model_2.predict(tfidf_test)
pred_score2=accuracy_score(y_test,prediction_2)
print(pred_score2*100)
```
